## Import data

```python
import numpy as np
import dill as pickle
import matplotlib.pyplot as plt

klambda = 1000
kdelta = 1000

run_ID = "Fasymp20_fixgaussian20_50GW"
pkl_fname = f'./Data/{run_ID}_Lookup_table_lambda_{klambda}_by_delta_{kdelta}_allangles.pkl'
with open(pkl_fname, 'rb') as data_file:
    data = pickle.load(data_file)

Q1 = data['Q1']
Q2 = data['Q2']
PD_Q1_delta = data['PD_Q1_delta']
PD_Q2_delta = data['PD_Q2_delta']
PD_Q1_lambda = data['PD_Q1_lambda']
PD_Q2_lambda = data['PD_Q2_lambda']

delta_array = data['delta array']
lambda_array = data['lambda array']
```

## Plot over angles

```python
## Select wavelength
i = 0

## Efficiency factors
fig1, (ax1,dummy,ax2) = plt.subplots(nrows=1, ncols=3, width_ratios=(1,0.1,1), figsize=(10,5))
dummy.axis('off')

ax1.plot((180/np.pi)*delta_array, Q1[i])
ax2.plot((180/np.pi)*delta_array, Q2[i])

ax1.set_xlim(xmin=(180/np.pi)*delta_array[0], xmax=(180/np.pi)*delta_array[-1])
ax2.set_xlim(xmin=(180/np.pi)*delta_array[0], xmax=(180/np.pi)*delta_array[-1])

ax1.set_xlabel("$\delta'$ (degrees)")
ax1.set_ylabel("$Q_{\\text{pr,1}'} (\delta')$")
ax2.set_xlabel("$\delta'$ (degrees)")
ax2.set_ylabel("$Q_{\\text{pr,2}'} (\delta')$")

fig1.suptitle(rf"Efficiency factors at $\lambda'=$ {lambda_array[i]:.2f}")

## Angular derivative
fig2, (ax3,dummy2,ax4) = plt.subplots(nrows=1, ncols=3, width_ratios=(1,0.1,1), figsize=(10,5))
dummy2.axis('off')

ax3.plot((180/np.pi)*delta_array, PD_Q1_delta[i])
ax4.plot((180/np.pi)*delta_array, PD_Q2_delta[i])

ax3.set_xlim(xmin=(180/np.pi)*delta_array[0], xmax=(180/np.pi)*delta_array[-1])
ax4.set_xlim(xmin=(180/np.pi)*delta_array[0], xmax=(180/np.pi)*delta_array[-1])

ax3.set_xlabel("$\delta'$ (degrees)")
ax3.set_ylabel(r"$\frac{\partial Q_{\text{pr,1}'}}{\partial \delta'} (\delta')$")
ax4.set_xlabel("$\delta'$ (degrees)")
ax4.set_ylabel(r"$\frac{\partial Q_{\text{pr,2}'}}{\partial \delta'} (\delta')$")

fig2.suptitle(rf"Angular derivatives at $\lambda'=$ {lambda_array[i]:.2f}")

## Wavelength derivative
fig3, (ax5,dummy3,ax6) = plt.subplots(nrows=1, ncols=3, width_ratios=(1,0.1,1), figsize=(10,5))
dummy3.axis('off')

ax5.plot((180/np.pi)*delta_array, PD_Q1_lambda[i])
ax6.plot((180/np.pi)*delta_array, PD_Q2_lambda[i])

ax5.set_xlim(xmin=(180/np.pi)*delta_array[0], xmax=(180/np.pi)*delta_array[-1])
ax6.set_xlim(xmin=(180/np.pi)*delta_array[0], xmax=(180/np.pi)*delta_array[-1])

ax5.set_xlabel("$\delta'$ (degrees)")
ax5.set_ylabel(r"$\frac{\partial Q_{\text{pr,1}'}}{\partial \lambda'} (\delta')$")
ax6.set_xlabel("$\delta'$ (degrees)")
ax6.set_ylabel(r"$\frac{\partial Q_{\text{pr,2}'}}{\partial \lambda'} (\delta')$")

fig3.suptitle(rf"Wavelength derivatives at $\lambda'=$ {lambda_array[i]:.2f}")

fig1.savefig("./Data/Q_vs_angle.pdf", bbox_inches="tight")
fig2.savefig("./Data/PDQPDdelta_vs_angle.pdf", bbox_inches="tight")
fig3.savefig("./Data/PDQPDlambda_vs_angle.pdf", bbox_inches="tight")
```

## Plot over wavelength

```python
## Select angle
j = 500

## Efficiency factors
fig1, (ax1,dummy,ax2) = plt.subplots(nrows=1, ncols=3, width_ratios=(1,0.1,1), figsize=(10,5))
dummy.axis('off')

ax1.plot(lambda_array, Q1[:,j])
ax2.plot(lambda_array, Q2[:,j])

ax1.set_xlim(xmin=lambda_array[0], xmax=lambda_array[-1])
ax2.set_xlim(xmin=lambda_array[0], xmax=lambda_array[-1])

ax1.set_xlabel("$\lambda'$")
ax1.set_ylabel("$Q_{\\text{pr,1}'} (\lambda')$")
ax2.set_xlabel("$\lambda'$")
ax2.set_ylabel("$Q_{\\text{pr,2}'} (\lambda')$")

fig1.suptitle(rf"Efficiency factors at $\delta'=$ {(180/np.pi)*delta_array[j]:.2f}$\degree$")

## Angular derivative
fig2, (ax3,dummy2,ax4) = plt.subplots(nrows=1, ncols=3, width_ratios=(1,0.1,1), figsize=(10,5))
dummy2.axis('off')

ax3.plot(lambda_array, PD_Q1_delta[:,j])
ax4.plot(lambda_array, PD_Q2_delta[:,j])

ax3.set_xlim(xmin=lambda_array[0], xmax=lambda_array[-1])
ax4.set_xlim(xmin=lambda_array[0], xmax=lambda_array[-1])

ax3.set_xlabel("$\lambda'$")
ax3.set_ylabel(r"$\frac{\partial Q_{\text{pr,1}'}}{\partial \delta'} (\lambda')$")
ax4.set_xlabel("$\lambda'$")
ax4.set_ylabel(r"$\frac{\partial Q_{\text{pr,2}'}}{\partial \delta'} (\lambda')$")

fig2.suptitle(rf"Efficiency factors at $\delta'=$ {(180/np.pi)*delta_array[j]:.2f}$\degree$")

## Wavelength derivative
fig3, (ax5,dummy3,ax6) = plt.subplots(nrows=1, ncols=3, width_ratios=(1,0.1,1), figsize=(10,5))
dummy3.axis('off')

ax5.plot(lambda_array, PD_Q1_lambda[:,j])
ax6.plot(lambda_array, PD_Q2_lambda[:,j])

ax5.set_xlim(xmin=lambda_array[0], xmax=lambda_array[-1])
ax6.set_xlim(xmin=lambda_array[0], xmax=lambda_array[-1])

ax5.set_xlabel("$\lambda'$")
ax5.set_ylabel(r"$\frac{\partial Q_{\text{pr,1}'}}{\partial \lambda'} (\lambda')$")
ax6.set_xlabel("$\lambda'$")
ax6.set_ylabel(r"$\frac{\partial Q_{\text{pr,2}'}}{\partial \lambda'} (\lambda')$")

fig3.suptitle(rf"Efficiency factors at $\delta'=$ {(180/np.pi)*delta_array[j]:.2f}$\degree$")

fig1.savefig("./Data/Q_vs_lambda.pdf", bbox_inches="tight")
fig2.savefig("./Data/PDQPDdelta_vs_lambda.pdf", bbox_inches="tight")
fig3.savefig("./Data/PDQPDlambda_vs_lambda.pdf", bbox_inches="tight")
```
