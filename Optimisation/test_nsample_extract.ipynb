### Sort the maxima found from each core

```python
### IMPORTS ###
import pickle
import sys
sys.path.append('../')

runID = "nsample_convergence_test"
num_cores = 5
maxfev = 5

pkl_fname = f'./Data/{runID}_maxfev{maxfev*num_cores}'
txt_fname = f'./Data/{runID}_maxfev{maxfev*num_cores}_curated.txt'

opt_FOMs = []
nsamples = []
for n in range(num_cores):
    try: 
        data_fname = pkl_fname + f"_process{n}.pkl"
        with open(data_fname, 'rb') as data_file:
            data = pickle.load(data_file)
    except FileNotFoundError:
        continue

    opt_FOMs.append(data["FOM"])
    nsamples.append(data["Sampling settings"]['n_sample'])
```

### Plotting convergence

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1)
ax.plot(nsamples, opt_FOMs, 'o')
ax.set(xlabel='Number of samples', ylabel='FOM', title='FOM vs Number of samples')
```
